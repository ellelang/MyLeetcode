```python
# Load the customer_data
customer_data = pd.read_csv(data_folder/'customer_data.csv')

# Load the app_purchases
app_purchases = pd.read_csv(data_folder/'inapp_purchases.csv')
```

```python
customer_data['reg_date'] = pd.to_datetime(customer_data['reg_date']).dt.strftime('%Y-%m-%d')
customer_data['reg_date'] 
```

```python
customer_data['uid'] = customer_data['uid'].astype(int)
customer_data['uid'] 
```

```python
uid_date_combined_data = app_purchases.merge(customer_data, left_on=['uid','date'], right_on=['uid','reg_date'], how='inner')
```

```python
# Examine the resutls
uid_date_combined_data.head(6)
```

```python
 purchase_data = uid_date_combined_data
```

```python
# Calculate the mean purchase price 
purchase_price_mean = purchase_data.price.agg('mean')

# Examine the output 
print(purchase_price_mean)

# Calculate the mean and median purchase price 
purchase_price_summary = purchase_data.price.agg(['mean', 'median'])

# Examine the output 
print(purchase_price_summary)

# Calculate the mean and median of price and age
purchase_summary = purchase_data.agg({'price': ['mean', 'median'], 'age': ['mean', 'median']})

# Examine the output 
print(purchase_summary)
```

```python
# Group the data 
grouped_purchase_data = purchase_data.groupby(by = ['device', 'gender'])
grouped_purchase_data.head(6)
```

```python
# Aggregate the data
purchase_summary = grouped_purchase_data.agg({'price': ['mean', 'median', 'std']})
# Examine the results
print(purchase_summary)
```
